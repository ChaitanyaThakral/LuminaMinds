# Exploratory Data Analysis: Suicide Watch
This notebook analyzes the psycholinguistic markers and vocabulary overlap for crisis detection.

```python
import sys
import os
sys.path.append(os.path.abspath('..'))

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import warnings
warnings.filterwarnings('ignore')

# Set up matplotlib style
plt.style.use('default')

from analytics.data_processing import (
    compute_post_lengths,
    compute_jaccard_similarity,
    run_psycholinguistic_analysis,
    readability_by_class
)
```

```python
DATASET_PATH = '../dataset/suicide_watch.csv'
df = pd.read_csv(DATASET_PATH).sample(20000, random_state=42)
df.head()
```

## 1. Jaccard Similarity

```python
jaccard = compute_jaccard_similarity(df)
print(f'Vocabulary Overlap (Jaccard Similarity): {jaccard:.4f}')
```

## 2. Psycholinguistic Markers (Mann-Whitney U)

```python
psych_df = run_psycholinguistic_analysis(df)
display(psych_df)
```

## 3. Flesch-Kincaid Readability

```python
fk_df = readability_by_class(df)
display(fk_df)
if hasattr(fk_df, 'attrs') and 'p_value' in fk_df.attrs:
    print(f"U-statistic: {fk_df.attrs['u_statistic']}, p-value: {fk_df.attrs['p_value']}, Cohen's d: {fk_df.attrs['cohens_d']}")
```
